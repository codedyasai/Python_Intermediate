## 불필요한 열 제거

```python
titanic = titanic.drop(['Cabin', 'WikiId', 'Name_wiki',
       'Age_wiki', 'Hometown', 'Boarded', 'Destination', 'Lifeboat', 'Body'], axis= 1)

titanic.head()
```

```python
# 수치데이터의 기술통계량
titanic.describe()
```

- 훈련 데이터(우리가갖고온게이거임)의 샘플 수: 891
- 훈련 데이터 내 생존률: 38.38%(Survived의 mean 값)


```python
# 문자데이터(범주형 데이터)의 통계량
titanic.describe(include='object')
```

- top은 가장 자주 나타나는 값
- freq는 가장 자주 나타나는 값에 대한 통계
- 훈련 데이터의 남성의 수: 577명(top의 sex가 male, freq 577이었음)
- 훈련 데이터중 가장 많은 승선항: S, 644명


```python
# 객실 등급(Pclass)에 따른 생존률 비교
## Pclass, Survived
titanic[['Pclass', 'Survived']].groupby(['Pclass'], as_index= False).mean().sort_values(by='Survived', ascending= False)
## groupby에 as_index를 False로 하면 Pclass를 index로 사용하지 않는다 라는 의미
## ascending: 오름차순
## as_index를 True로 하면 Pclass를 index로 사용한다면 의미
```

- 객실 등급이 높을수록 생존율이 좋다.

```python
# 성별(Sex)에 따른 생존율 비교
titanic[['Sex', 'Survived']].groupby(['Sex'], as_index= False).mean().sort_values(by= 'Survived', ascending=False)
```

여성의 생존율이 남성보다 높다.

```python
# 함께 승선한 형제자매 배우자 수(SibSp)에 따른 생존율 비교
```

```python
# 함께 승선한 부모와 자녀수(Parch)에 따른 생존율 비교
titanic[['Parch', 'Survived']].groupby(['Parch'], as_index= False).mean().sort_values(by= 'Survived', ascending=False)
```

- 동반 가족수가 적을수록 생존율이 높다

```python
# 생존여부에 따른 연령(Age)의 분포 - 시각화(histogram)
import seaborn as sns
# 열(col)을 생존 여부로 나눔
g = sns.FacetGrid(titanic, col = 'Survived')
# 히스토그램으로 시각화, 연령의 분포를 확인, 히스토그램의 bin을 20으로 설정
g.map(plt.hist, 'Age', bins=20)
plt.show()
```

```python
# 객실 등급(Pclass)과 생존 여부(Survived)에 따른 연령(Age) 분포
g = sns.FacetGrid(titanic, col = 'Survived', row = 'Pclass', hue= 'Pclass', height= 2.2, aspect= 1.6)
g.map(plt.hist, 'Age', bins=20, alpha= 0.5)
g.add_legend()
plt.show()
```

- 객실 등급이 3등급인 경우가 사람이 제일 많고 생존율이 낮다.
- 객실 등급이 2등급인 경우 유아는 대부분 생존했다.
- 객실 등급이 1등급인 경우 생존율이 비교적 높다.

```python
# 승선항(Embarked)와 객실 등급(Pclass)에 따른 생존율(Survived) 시각화
g = sns.FacetGrid(titanic, row= 'Embarked', height= 2.2, aspect = 1.6)
# sns.pointplot(data, x, y, hue)
# palette 옵션: 전체를 하나의 색깔로 지정하는 것이 아니라 각기 다른 색으로 설정하기 위한 옵션
# order는 x축 순서
g.map(sns.pointplot, 'Pclass', 'Survived', 'Sex', palette= 'deep', order=[1,2,3], hue_order= ['male', 'female'])
g.add_legend()

#수직 막대는 신뢰구간
```

```python
# 성별(Sex) 변수를 숫자형으로 변경하는 작업

titanic['Sex'] = titanic['Sex'].map({'female':1, 'male':0}).astype(int)

titanic['Sex']
```
