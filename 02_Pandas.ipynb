# Pandas
- R의 자료구조 dataframe, matrix(2차원 배열)
_ 데이터 분석에서 데이터의 핸들링을 할 때 필수 라이브러리
- 자료형: Series와 Dataframe
- 표(table) 형태: 행(관측치, 레코드), 열(속성, 피처)
- 시리즈(Series) = 인덱스(index) + 값(value)

## 1.1 Series
- 1차원 배열과 비슷
- 구성: index와 value
- 생성: pd.Series(data, index)

```python
# 모듈 로딩
import numpy as np
import pandas as pd
```

```python
price = pd.Series([4000, 3000, 5000, 2000])
price
```

```python
price.index
```

```python
price.values
```

```python
fruits = pd.Series([4000, 3000, 5000, 2000],
                   index=['aaa','bbb', 'ccc', 'ddd']
                   )
fruits
```

```python
fruits['aaa']
```

```python
fruits['aaa': 'ccc']
```

```python
# dict를 이용한 시리즈 생성
city_dict = {'Seoul': 82, 'Busan': 90, 'Incheon':84, 'Daejeon': 42}
city_dict
```

```python
city = pd.Series(city_dict)
city
```

```python
city.index
```

```python
city.values
```

```python
city['Incheon']
```

```python
obj = pd.Series([4,7,-5,3])
obj
```

```python
obj2 = pd.Series([4,7,-5,3], index=['d','b','a','c'])
obj2
```

```python
obj2['c']
```

```python
obj2[['c','a','d']]
```

```python
obj2[obj2 > 0]
```

```python
# in and not in
'c' in obj2
```

## DataFrame
- 행과 열로 구성
- 행(관측치), 열(속성, 피처)
- pd.DataFrame(data.index,columns)
- 데이터분석에서 가장 기본이 되는 자료구조
- 여러개의 Series로 구성

```python
city_dict = {'name':['aaa', 'bbb', 'ccc'], 'age':[13,24,19]}
city_dict
```

```python
city = pd.DataFrame(city_dict)
city
```

```python
city.index
```

```python
city.columns
```

```python
city[0, :]
```

```python
data = {
    "2015": [9904312, 3448737, 2890451, 2466052],
    "2010": [9631482, 3393191, 2632035, 2431774],
    "2005": [9762546, 3512547, 2517680, 2456016],
    "2000": [9853972, 3655437, 2466338, 2473990],
    "지역": ["수도권", "경상권", "수도권", "경상권"],
    "2010-2015 증가율": [0.0283, 0.0163, 0.0982, 0.0141]
}
columns = ["지역", "2015", "2010", "2005", "2000", "2010-2015 증가율"]
index = ["서울", "부산", "인천", "대구"]
df = pd.DataFrame(data, index=index, columns=columns)
df
```

```python
df.values
```

```python
df.columns
```

```python
df.index
```

```python
# 열 추가
df['2005-2010 증가율'] = ((df['2010']-df['2005'])/ df['2005']*100).round(2)
df
```

```python
# 열 삭제
del df['2010-2015 증가율']
df
```

```python
# 열 인덱싱: 시리즈로 반환
df['지역']
```

```python
df[['2010','2015']]
```

```python
df[['지역']]
```

```python
type(df[['지역']])
```

```python
type(df['지역'])
```
