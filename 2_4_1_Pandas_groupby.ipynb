# Groupby : 같은 값을 하나로 묶어 통계 또는 집계결과를 얻기 위한 기능

```python
df = pd.DataFrame({
    'city':['부산','부산','부산','부산','서울','서울','서울'],
    'fruits':['apple','orange','banana','banana','apple','apple','banana'],
    'price':[100,200,250,300,150,200,400],
    'quantity':[1,2,3,4,5,6,7]
})
df
```

## get_group(): 그룹 안에 데이터를 확인

```python
df.groupby('city').mean()
```

```python
df.groupby('city').get_group('부산')
```

size(): 각 그룹의 사이즈 확인

```python
df.groupby('city').size()
```

```python
df.groupby('city').size()['부산']
```

## agg(): Aggregation의 약자

```python
df.groupby('city').agg(np.mean)
```

```python
# 가격 평균과 수량은 합계를 동시에 계산
def my_mean(s):
    return np.mean(s)

df.groupby('city').agg({'price': my_mean, 'quantity': sum})
```

### 대륙과 국가별 음주 소비량을 분석한 데이터셋
- 각각의 피처들간의 상관관계 파악
- 아프리카와 유럽간의 맥주 소비량 차이 검정

```python
drinks = pd.read_csv('drinks.csv')
drinks.info()
```

```python
drinks.describe()
```

```python
# 각 피처들간의 상관관계 파악 - pearson 상관계수
```

```python
# 대륙(continent)의 결측값을 대체값(OT)로 채우는 작업
## 결측여부 확인: isnull().sum()
## 결측값 대체: fillna(value)
drinks['continent'] = drinks['continent'].fillna('OT')
drinks.head(10)
```

```python
# 대륙별로 맥주 소비량의 최대, 최소, 평균을 추출
res = drinks.groupby('continent')['beer_servings'].agg(['max', 'min', 'mean'])
res
```

```python
# idxmax(): Series 객체에서 값이 가장 큰 index를 반환하는 메서드
# 평균 맥주 소비량이 가장 큰 대륙은?
res = drinks.groupby('continent')['beer_servings'].mean().idxmax()
res
```

```python
# t-test: 차이검정
# 아프리카와 유럽 간의 맥주 소비량 차이를 검정
from scipy import stats # import scipy.stats as stats 로 쓰는 사람들도 있다
africa = drinks[drinks['continent']== 'AF']
europe = drinks[drinks['continent']== 'EU']
```

```python
# 정규분포 - 정규성
# shapiro test - 정규성 검증, p-value > 0.05면 정규분포다.
# 귀무가설: 차이가 없다, 일반적인 가설. p-value가 더 크면 채택
# 대립가설: 연구자의 가설, 내가 주장하고 싶은 가설. p-value 가 더 작으면 채택

sha_a, sha_pvalue = stats.shapiro(africa['beer_servings'])
print(sha_a, sha_pvalue)
```

```python
# 등분산: 분산의 동질성 검사, levene test
e, p = stats.levene(africa['beer_servings'], europe['beer_servings'])
print(e, p)

if p > 0.05:
    print('등분산이다,')
else:
    print('등분산이 아니다.')
```

# 카이제곱 검정(Chi square test)
- 두 데이터가 통계적으로 같은지 다른지 검정하는 방법
- 카이제곱 검정에는 두 가지 형태가 있으며, 카이제곱 통계량과 분포를 사용하지만 목적에 따라 달라짐
- 적합도 검정: 관찰된 값이 기대값과 같은지 다른지 검정
- 동질성 검정: 두 집단의 분포가 동일한지 검정
- 독립성 검정: 두 개 이상의 변수가 서로 독립인지 검정

```python
data1 = [4,6,17,16,8,9]
data2 = [10,10,10,10,10,10]

# 집단의 대표값
print(np.mean(data1))
print(np.mean(data2))
```

```python
# 카이제곱 검정: 두 데이터가 통계적으로 같은지를 검정

chis = stats.chisquare(data1, data2)
chis
```

# T-검정: 차이 검정
## 단일표본 t검정(1개의 그룹): 연속형 변수를 대상으로
  - ttest_1samp()
## 독립표본 t검정(서로 다른 2개의 그룹)
  - ttest_ind()
## 대응표본 t검정(같은 그룹)
  - ttest_rel()

  

```python
# 단일표본 t검정: 1개의 그룹
# 일반적으로 청소년의 신장은 175이다.
np.random.seed(100)
# 평균 0, 표준편차 5인 20개의 난수 발생
height = [175 + np.random.normal(0,5) for a in range(20)]
print(height)
```

```python
result = stats.ttest_1samp(height, 175)
print(result)
```

- p-value > 0.05, 귀무가설 채택.

```python
# 독립표본 r 검정(서로 다른 2개의 집단을 이용)
group1 = [170 + np.random.normal(0,5) for a in range(20)]
group2 = [175 + np.random.normal(0,5) for a in range(20)]

print(group1)
print(group2)
```

```python
# 등분산 가정 (분산이 같을 경우)
result1 = stats.ttest_ind(group1, group2)
print(result1)

# 등분산 가정을 하지 않을 경우 (분산이 다른 경우)
result2 = stats.ttest_ind(group1, group2, equal_var= False)
print(result2)
```

```python
# 대응표본 t검정(같은 그룹을 대상으로, 반드시 샘플데이터의 수가 동일해야 한다.)

before = [60+np.random.normal(0,5) for a in range(20)]
print(before)


after = [ 60 + w * np.random.normal(0.99,0.02) for w in range(20)]
print(after)
```

```python
res = stats.ttest_rel(before, after)
print(res)
```

```python
# 아프리카와 유럽간의 맥주 소비량 차이를 검정

a = stats.ttest_ind(africa['beer_servings'], europe['beer_servings'])
b = stats.ttest_ind(africa['beer_servings'], europe['beer_servings'], equal_var= False)
print(a)
print(b)
```
