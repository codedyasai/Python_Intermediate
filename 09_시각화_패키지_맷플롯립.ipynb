## 범례(legend)
- 하나의 차트에 여러개의 그래프를 그릴 경우, 각 그래프가 뜻하는 것을 표시하는 것
- legend()
- loc: best(0)
- loc: upper right(1)
- loc: upper left(2)
- loc: lower left(3)
- loc: lower right(4)
- loc: right(5)
- loc: center left(6)
- loc: center right(7)
- loc: lower center(8)
- loc: upper center(9)
- loc: center(10)

```python
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
s = np.sin(x)
plt.title('범례를 가진 그래프')
plt.plot(x,c, ls= '--', label = 'cosine')
plt.plot(x,s, ls = ':', label = 'sine')
plt.legend(loc=2)
plt.show()
```

## 그림 구조
- Figure 객체: 그림 객체. 그림이 그려지는 캔버스. 도화지.
- axes 객체: Figure 객체는 하나 이상의 Axes 객체
- axis 객체: 축

```python
fig = plt.figure(figsize=(10,2))
plt.title('fig 객체를 이용한 도화지 크기 조정')
plt.plot(np.random.randn(100))

plt.show()
```

**Axes객체와 subplot()**
- subplot(2,1,1) = 도화지를 2행 1열로 나누고 첫번째 행에, 라는 뜻

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

ax1 = plt.subplot(2,1,1) # 첫번째 축
plt.plot(x1, y1, 'yo-')
plt.title('첫번째 subplot')
plt.ylabel('A')

ax2 = plt.subplot(2,1,2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time')
plt.ylabel('B')

# 플롯간의 간격을 자동으로 맞춰주는 메서드
plt.tight_layout()
plt.show()
```

```python
plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title('axes 1')

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title('axes 2')

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title('axes 3')

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title('axes 4')

plt.tight_layout()
plt.show()
```

**Axis 객체와 축**
- 하나의 Axes에 두 개 이상의 Axis 객체를 가진다.
- twinx() 메서드를 이용하여 복수개의 축을 설정한다.

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()

ax0.set_title('2개의 y축을 가진 하나의 figure객체')
ax0.plot([10, 5, 2, 9, 7], 'r-', label='y0')
ax0.set_ylabel('y0')
ax0.grid(True)

ax1.plot([100, 200, 220, 180, 120], 'g:', label='y1')
ax1.set_ylabel('y1')
ax1.grid(True)

ax0.set_xlabel('공유하는 x축')
plt.show()
```

```python
fig, ax = plt.subplots(3,4, figsize=(15,10))

for row in range(3):
    for col in range(4):
        ax[row, col].plot(np.random.randn(10))
        ax[row, col].set_title(f"{row+1}행 {col+1}열")
        ax[row, col].set_xlabel(f"{row+1}행")
        ax[row, col].set_ylabel(f"{col+1}열", rotation=0)

fig.tight_layout()
plt.show()
```

```python
y = [2,3,1]
x = np.arange(len(y))

xlabel = ['가', '나', '다']

plt.title('bar chart')
plt.bar(x,y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel('가나다')
plt.ylabel('빈도수')

plt.show()
```

```python
y = [2,3,1]
x = np.arange(len(y))

xlabel = ['가', '나', '다']

plt.title('bar chart')
plt.barh(x,y, alpha = 0.4)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel('가나다')
plt.ylabel('빈도수')

plt.show()
```

## 파이 차트(Pie Chart)
- 카테고리별값이 상대적인 비교를 할 때 사용
- pie()
- 파이 차트를 그릴 때 원의 형태를 유지하도록 plt.axis('equal') 명령

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)

# title
plt.title('Pie Chart')


# pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='1.1f%%', shadow=True, startangle=90)

# pie chart 사용시 실행
plt.axis('equal')

plt.show()
```

## 스캐터 플롯(scatter plot)
- 산점도(산포도): 데이터의 분포를 파악하는 시각화도구


```python
np.random.seed(0)

x = np.random.normal(0,1, 100)
y = np.random.normal(0, 1, 100)

plt.title('Scatter Plot')

plt.scatter(x,y)

plt.show()
```

```python
# 2차원이 아니라 3차원 또는 4차원인 경우에는 점 하나의 크기 혹은 색깔을 이용해서 시각화한다.
# 버블 차트(bubble chart)라고 부른다.
# s: 크기, c: 색상
N = 30
np.random.seed(0)
x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15 * np.random.rand(N)) ** 2

plt.title('Bubble Chart')

plt.scatter(x, y1, c=y2, s=y3)

plt.show()
```
