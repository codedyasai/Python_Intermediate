```python
cctv_seoul.drop('Unnamed: 0', axis= 1, inplace= True)
```

```python
# 컬럼명 변경: 구분 -> 구별
cctv_seoul.rename(columns = {cctv_seoul.columns[0]: '구별'}, inplace= True)
```

```python
cctv = pd.read_excel('seoul_cctv.xlsx')
cctv.head()
```

```python
# 컬럼명 변경: 기관명 -> 구별
cctv.rename(columns= {cctv.columns[0]:'구별'}, inplace= True)
cctv.head()
```

```python
cctv.info()
```

```python
cctv.isnull().sum()
```

```python
cctv = cctv.fillna(0)
```

```python
# 서울시 인구현황 데이터 로딩
pop = pd.read_excel('seoul_popu.xls')
pop.head()
```

```python
# header = 2, 세번째 줄부터 읽어오는 옵션
# parse_cols = 'B, D, G, J, N' 열만 읽어오는 옵션
pop = pd.read_excel('seoul_popu.xls', header= 2, usecols='B,D,G,J,N')
pop.head()
```

```python
# 컬럼명 변경
pop.rename(columns={pop.columns[0]: '구별',
                    pop.columns[1]: '인구수',
                    pop.columns[2]: '한국인',
                    pop.columns[3]: '외국인',
                    pop.columns[4]: '고령자'},inplace= True)

pop.head()
```

```python
cctv.head()
```

```python
cctv.columns
```

```python
# 2016년 이전 :  '2011년 이전', '2012년', '2013년', '2014년', '2015년', '2016년'
# 파생변수: 기존의 변수를 이용하여 새로운 변수를 생성하는 것
cctv['2016년 이전'] = cctv['2011년 이전'] + cctv['2012년'] + cctv['2013년'] + cctv['2014년'] + cctv['2015년'] + cctv['2016년']
```

```python
cctv.columns
```

```python
cctv = cctv.drop(['2011년 이전', '2012년', '2013년', '2014년', '2015년', '2016년'], axis=1)
cctv.head()
```

```python
# cctv 적게 설치된 구를 확인
# 소계를 기준으로 오름차순

cctv.sort_values(by= '소계').head(3)
```

```python
# 최근 2년간 cctv 증가율 계산: 최근 증가율

cctv['최근증가율'] = (cctv['2017년']+cctv['2018년'])/ cctv['2016년 이전'] * 100
cctv.sort_values(by='최근증가율', ascending= False).head()
```

- 최근 2년간 cctv가 많이 증가한 구는 광진구, 영등포구, 성동구, 금천구, 중구

```python
# 합계 행 제거
pop.drop([0], inplace= True)
pop.head()
```

```python
pop['구별'].unique()
```

```python
# 구별 외국인 비율과 고령자 비율 계산하여 각각 파생변수에 담는다
pop['외국인비율'] = pop['외국인'] / pop['인구수'] * 100
pop['고령자비율'] = pop['고령자'] / pop['인구수'] * 100
pop.head()
```

```python
# 인구수가 많은 순으로 정렬
pop.sort_values(by= '인구수', ascending= False).head()
```

```python
# 외국인 비율이 높은 순으로 정렬
pop.sort_values(by= '외국인비율', ascending= False).head()
```

```python
# 고령자 수가 많은 순으로 정렬
pop.sort_values(by= '고령자', ascending= False).head()
```

```python
# 고령자 비율이 높은 순으로 정렬
pop.sort_values(by= '고령자비율', ascending= False).head()
```

# CCTV데이터와 인구현황 데이터를 합쳐 분석

```python
df = pd.merge(cctv, pop, on='구별')
df.head()
```

- 의미없는 컬럼 또는 행을 삭제하는 작업
- 행 방향: drop, 열 방향 del

```python
# '구별' 컬럼을 인덱스로 설정
df.set_index('구별', inplace= True)
df.head()
```
